```python
import os
import numpy as np
```

```python
kitti_dir = '../DETRAC/'
kit_main = 'DETRAC/'
kitti_subdir = os.listdir(kitti_dir)
img_dirs = os.listdir(os.path.join(kitti_dir, kitti_subdir[0]))
full_img_dirs = []
for dir in img_dirs:
    # full_img_dirs.append(os.path.join(kitti_dir,
    #                     os.path.join(kitti_subdir[0], dir)))
    full_img_dirs.append(kit_main + kitti_subdir[0] + '/' + dir)
full_img_dirs
```

```python
train_files = []
for dir in full_img_dirs:
    files = os.listdir('../'+dir)
    for imgfile in files:
        # fullpath = os.path.join(dir, imgfile)
        fullpath = dir + '/' + imgfile
        train_files.append(fullpath)
train_files
```

```python
np.savetxt('detrac_test.train', train_files, fmt='%s')
```

```python
img_dict = {}
for img_fld in dimgDir:
    img_dict[img_fld] = os.listdir(os.path.join(detrac_img_dir, img_fld))
img_dict.keys()
```

```python
lbl_dict = {}
for lbl_fld in dlblDir:
    lbl_dict[lbl_fld] = os.listdir(os.path.join(detrac_lbl_dir, lbl_fld))
lbl_dict.keys()
```

```python
img_dict['MVI_20012'][0][:-4]
```

```python
issues_dict = {}
for folder in img_dict.keys():
    print(folder)
    img_files_len = len(img_dict[folder])
    lbl_files_len = len(lbl_dict[folder])
    if (img_files_len != lbl_files_len):
        issues_dict[folder] = []
        for idx in range(len(img_dict[folder])):
            try:
                if img_dict[folder][idx][:-4] != lbl_dict[folder][idx][:-4]:
                    print('y')
                    # issues_dict[folder].append(img_dict[folder][idx])
            except:
                print('how come?')    
                issues_dict[folder].append(img_dict[folder][idx])
                
    # assert img_files_len==lbl_files_len, folder
```

```python
issues2_dict = {}
for folder in img_dict.keys():
    print(folder)
    img_files_len2 = len(img_dict[folder])
    lbl_files_len2 = len(lbl_dict[folder])
    if (img_files_len2 != lbl_files_len2):
        issues2_dict[folder] = []
        temp_img = img_dict[folder][:][:-4]
        temp_lbl = lbl_dict[folder][:][:-4]
        diff = list(set(temp_img) - set(temp_lbl))
        issues2_dict[folder] = diff
        print(sorted(diff))
```
